```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
```

# Plots de antebrazo

```python
# --- Cargar archivo ---
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\PUCP\ISB\Lab3\Mediciones Antebrazo\Antebrazo Basal 1.txt",
                 sep="\t", comment="#", header=None)

df.columns = ["nSeq", "I1", "I2", "O1", "O2", "A1", "extra"]

# Extraer señal
emg = df["A1"].values
fs = 1000  # Hz Extraido de la documentación
N = len(emg)
t = np.arange(N) / fs

# --- FFT ---
freqs = np.fft.rfftfreq(N, 1/fs)
fft_vals = np.fft.rfft(emg)
fft_magnitude = np.abs(fft_vals) / N   # normalizar

# Convertir a dB (añado un epsilon para evitar log(0))
fft_dB = 20 * np.log10(fft_magnitude + 1e-12)

# FFFT con Welch method 
freqs_welch, psd_welch = signal.welch(emg, fs, nperseg=2048, 
                                      window='hann')
                                      
psd_dB_welch = 10 * np.log10(psd_welch + 1e-12)  # PSD usa 10*log10

# ---------- Plots ----------
plt.figure(figsize=(15, 4))

# Señal cruda
plt.subplot(131)
plt.plot(t, emg)
plt.title("Señal EMG antebrazo basal")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [u.a.]")
plt.grid(True)

# FFT señal cruda
plt.subplot(132)
plt.plot(freqs, fft_dB)
plt.title("Espectro EMG (FFT)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud [dB]")
plt.xlim(0, fs/2)  # Nyquist
plt.grid(True)

# FFT con Welch
plt.subplot(133)
plt.plot(freqs_welch, psd_dB_welch, 'purple', linewidth=2)
plt.title("FFT Método de Welch")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("PSD [dB/Hz]")
plt.xlim(0, 300)
plt.grid(True, alpha=0.3)

plt.savefig("AntebrazoBasal.png")
plt.show()
```
